Executing content of `colab_rag.ipynb`:

````python
import nbformat
from IPython.display import display, Markdown

notebook_path = 'colab_rag.ipynb'

if os.path.exists(notebook_path):
    with open(notebook_path, 'r', encoding='utf-8') as f:
        notebook_content = nbformat.read(f, as_version=4)

    print(f"Generating cells from {notebook_path}...")
    for cell in notebook_content.cells:
        if cell.cell_type == 'code':
            # For code cells, generate a new code cell
            display(Markdown(f"```python\n{cell.source}\n```"))
            exec(cell.source) # Execute the code directly
        elif cell.cell_type == 'markdown':
            # For markdown cells, generate a new markdown cell
            display(Markdown(cell.source))
else:
    print(f"Error: The file '{notebook_path}' was not found.")
````
